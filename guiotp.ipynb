```python
import smtplib
from tkinter import *
from tkinter import messagebox
import random
import string

N = 6 
res = ''.join(random.choices(string.ascii_uppercase + string.ascii_lowercase + string.digits, k = N))

def send_message():
    
    f = open("D:\pass.txt", "r")
    x=f.read()

    
    
    sender_email = "mailtests1009@gmail.com" 
    
    sender_password = x
    
    server = smtplib.SMTP('smtp.gmail.com',587)
    
    server.starttls()
    
    server.login(sender_email,sender_password)
    
    print("Login successful")
    
    mylabel = Label(app, text="An OTP has been sent to your Email id")
    mylabel.pack()
    
    address_info = address.get()
    
    email_body_info = """\
    Thakur Testing
    
    Subject:""

    Your OTP for email verification is : """ + str(res)
    
    server.sendmail(sender_email,address_info,email_body_info)
   
    
def verify():
    if(ver_entry.get()==str(res)):
            messagebox.showinfo("Success", "Verification Successfull")
    else:
            messagebox.showinfo("Sorry", "Verification  not Successfull")
    
        
    
    
app = Tk()

app.geometry("500x500")

app.title("Thakur OTP Testing")

heading = Label(text="Welcome buddy",bg="yellow",fg="black",font="10",width="500",height="3")

heading.pack()

address_field = Label(text="Recipient Address :")

address_field.place(x=15,y=70)

address = StringVar()


address_entry = Entry(textvariable=address,width="30")


address_entry.place(x=15,y=100)

button = Button(app,text="Send OTP",command=send_message,width="30",height="2",bg="grey",fg="white")

button.place(x=15,y=150)

ver_field = Label(text="Enter OTP:")
ver_field.place(x=15,y=200)
ver = StringVar()
ver_entry = Entry(textvariable=ver,width="30")

ver_entry.place(x=15,y=250)
b1 = Button(app,text="verify otp",command=verify,width="30",height="2",bg="grey",fg="white")
b1.place(x=15,y=300)


mainloop()
#entry_text.set(entry_text.get()[:5])


```
